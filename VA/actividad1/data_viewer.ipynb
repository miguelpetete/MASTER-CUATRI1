```python
import random
import cv2 as cv
import matplotlib.pyplot as plt
import shutil
import os
```

```python
path_to_dataset = '/Users/miguelangelnavarroarenas/.cache/kagglehub/datasets/soumikrakshit/dark-face-dataset/versions/1'
```

```python
# Get 4 random images from the database
# Those images are stored as directories in an array
# Also, it returns the labels from the dataset if they are needed
def get_images():
    images = []
    labels = []
    for i in range(4): 
        rand_num = random.randint(1, 6000)
        images.append(path_to_dataset + f'/image/{rand_num}.png')
        labels.append(path_to_dataset + f'/label/{rand_num}.txt')
    return images, labels
```

```python
rand_images, rand_labels = get_images()
```

```python
print("IMAGES: ", rand_images)
```

```python
print("LABELS: ", rand_labels)
```

```python
def show_images_in_notebook(images):
    plt.figure(figsize=(10, 10))  # Tamao de la figura
    for i, img_path in enumerate(images):
        # Cargar la imagen en color
        img = cv.imread(img_path)
        if img is not None:
            # Convertir la imagen de BGR a RGB para mostrarla correctamente en matplotlib
            img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
            # Aadir la imagen a la cuadrcula
            plt.subplot(2, 2, i+1)
            plt.imshow(img_rgb)
            plt.title(f"Image {i+1}")
            plt.axis("off")  # Ocultar los ejes
        else:
            print(f"Error loading image at {img_path}")
    plt.show()

# Asumiendo que 'rand_images' contiene las rutas de las imgenes seleccionadas
show_images_in_notebook(rand_images)
```

```python
#COPIAREMOS LAS IMGENES GENERADAS EN ESTE DIRECTORIO PARA AS TENERLAS A MANO Y NO PERDERLAS
# Directorio de destino donde quieres almacenar las imgenes y etiquetas
dest_dir_images = "./imagenes_proyecto"
dest_dir_labels = "./etiquetas_proyecto"

# Asegrate de que los directorios de destino existen. Si no, cralos.
os.makedirs(dest_dir_images, exist_ok=True)
os.makedirs(dest_dir_labels, exist_ok=True)

# Listas con las rutas de las imgenes y etiquetas
# Supn que `image_paths` contiene las rutas de las imgenes seleccionadas
# Supn que `label_paths` contiene las rutas de las etiquetas correspondientes

# Copiar imgenes
for img_path in rand_images:
    try:
        shutil.copy(img_path, dest_dir_images)
        print(f"Imagen copiada: {img_path}")
    except Exception as e:
        print(f"Error al copiar la imagen {img_path}: {e}")

# Copiar etiquetas
for label_path in rand_labels:
    try:
        shutil.copy(label_path, dest_dir_labels)
        print(f"Etiqueta copiada: {label_path}")
    except Exception as e:
        print(f"Error al copiar la etiqueta {label_path}: {e}")
```
